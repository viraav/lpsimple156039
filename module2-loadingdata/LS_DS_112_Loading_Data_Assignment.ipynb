```
# Get the current list of columns
cols = df.columns.tolist()

# Remove the 'Full Name' column from the list
cols.remove('Full Name')

# Find the index of the 'Last Name' column
last_name_index = cols.index('Last Name')

# Insert 'Full Name' after 'Last Name'
cols.insert(last_name_index + 1, 'Full Name')

# Reindex the DataFrame with the new column order
df = df[cols]

display(df.head())
```

```
df['Full Name'] = df['Full Name'].str.title()
display(df.head())
```

```
#inserting a new column named â€œGMVâ€ after Seller Discount in the table
df['GMV'] = df['Transaction Amount'] - df['Seller Discount']
display(df.head())
```

```
display(df.head(2))
```

```
# Define the list of categories to include in "Group 1" and "Group 2"
group1_categories = ['Home/ Food', 'Sports/ Hobi/ Otomotif', 'Fashion', 'Beauty/ Health']
group2_categories = ['Gadget/ Komputer', 'Elektronik', 'Service/ Mokado', 'Babies/ Kids']

# Create the new column "Group 1" and "Group 2" and initialize with None or NaN
df['Group 1'] = None
df['Group 2'] = None

# Use boolean indexing to select rows where 'Meta Category' is in the list and assign the value from 'Meta Category'
df.loc[df['Meta Category'].isin(group1_categories), 'Group 1'] = df['Meta Category']
df.loc[df['Meta Category'].isin(group2_categories), 'Group 2'] = df['Meta Category']

# Display the head of the DataFrame to show the new column
display(df.head())
```

```
# Convert 'Paid Date' to datetime objects
df['Paid Date'] = pd.to_datetime(df['Paid Date'], format='%Y%m%d')

# Display the head of the DataFrame to show the new column
display(df.head())
```

```
total_gmv = df['GMV'].sum()
print(f"The total GMV is: {total_gmv}")
```

```
# Group by month and count the number of transactions
monthly_counts = df.groupby(df['Paid Date'].dt.to_period('M')).size().reset_index(name='Number of Transactions')

display(monthly_counts)
```
